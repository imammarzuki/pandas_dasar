```python
import pandas as pd

df = pd.read_csv("drive/MyDrive/pandas_dasar/Data/jamesbond.csv", index_col="Film")
df.sort_index(inplace=True)
```

```python
kond = df["Actor"] == "Sean Connery"
df[kond]
```

```python
kond = df["Actor"] == "Sean Connery"
df.where(kond)
```

```python
kond1 = df["Actor"] == "Sean Connery"
kond2 = df["Year"] == 1971
df.where(kond1 & kond2)
```
