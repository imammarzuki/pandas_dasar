Menggunakan Method to_datetime()

```python
df = pd.read_csv("employees.csv")
```

```python
df.info()
```

```python
df["Start Date"] = pd.to_datetime(df["Start Date"])
df["Last Login Time"] = pd.to_datetime(df["Last Login Time"])
```

```python
df.info()
```

```python
df["Last Login Time"]
```

```python
df["Gender"] = df["Gender"].astype("category")
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Team"] = df["Team"].astype("category")
```

```python
df.info()
```

Data Filtering

```python
df[df["Gender"]=="Female"]
```

```python
kondisi = df["Gender"]=="Female"
df[kondisi]
```

```python
kondisi = df["Team"] != "Marketing"
df[kondisi]
```

```python
kondisi1 = df["Gender"] == "Female"
kondisi2 = df["Team"] != "Marketing"
df[kondisi1 & kondisi2]
```

```python
kondisi1 = df["Senior Management"]
kondisi2 = df["Start Date"] < '1990-01-01'
df[kondisi1 | kondisi2]
```

Menggunakan isin()

```python
kond1 = df["Team"] == "Marketing"
kond2 = df["Team"] == "Finance"
kond3 = df["Team"] == "Sales"

df[kond1 | kond2 | kond3]
```

```python
kond = df["Team"].isin(["Marketing", "Finance", "Sales"])

df[kond]
```

Method isnull() dan notnull()

```python
kond1 = df["Team"].isnull()

df[kond1]
```

```python
kond1 = df["Team"].notnull()

df[kond1]
```

Mehtod between()

```python
kond1 = df["Salary"].between(60000, 70000)
df[kond1]
```
